## Mispriced diamonds

Insert data into R

```r
MyData <- read.csv(file="./datasets/Mispriced-Diamonds.csv")
```

Install ggplot library

```r
library(ggplot2)
```

Print plot

```r
ggplot(data=MyData[MyData$carat<2.5,], aes(x=carat, y=price, colour=clarity)) +
    geom_point(alpha=0.1) +
    geom_smooth()
```
